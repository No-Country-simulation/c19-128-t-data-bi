# Balanceo de datos con respecto a la columna Clase

```python
conteo_clases = data['Class'].value_counts()
print(conteo_clases)
```

```python
import matplotlib.pyplot as plt

conteo_clases = {
    0: 284315,
    1: 492
}

# Extraer las clases y los conteos
clases = list(conteo_clases.keys())
conteos = list(conteo_clases.values())

# Graficar los datos
plt.bar(clases, conteos, color=['blue', 'green'])  # Colores opcionales
plt.xlabel('Clase')
plt.ylabel('Cantidad')
plt.title('Conteo de Clases')
plt.xticks(clases)  # Asegura que los ticks del eje x correspondan a las clases
plt.show()
```

#### Solo 492(0.172%) de las transacciones son fraudulentas. Eso significa que hay un gran desbalanceo con respecto a la variable Clase.

# Continuamos explorando la data

### Densidad de tiempo para las clases de transacciones de tarjetas de crÃ©dito.

```python
import matplotlib.pyplot as plt
import seaborn as sns

clase_0 = data.loc[data['Class'] == 0]['Time']
clase_1 = data.loc[data['Class'] == 1]['Time']

# Graficamos

sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
sns.kdeplot(clase_0, label='No fraude', shade=True)
sns.kdeplot(clase_1, label='Fraude', shade=True)
plt.title('Tiempo para las clases de transacciones')
plt.xlabel('Time [s]')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

#### Podemos observar que las transacciones fraudulentas tienen una distribuciÃ³n mÃ¡s uniforme que las transacciones vÃ¡lidas, se distribuyen mÃ¡s equitativamente en el tiempo, incluidos los bajos tiempos de transacciÃ³n real en la zona horaria de Europa.
