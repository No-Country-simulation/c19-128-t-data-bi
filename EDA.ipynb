# Balanceo de datos con respecto a la columna Clase

```python
conteo_clases = data['Class'].value_counts()
print(conteo_clases)
```

```python
import matplotlib.pyplot as plt

conteo_clases = {
    0: 284315,
    1: 492
}

# Extraer las clases y los conteos
clases = list(conteo_clases.keys())
conteos = list(conteo_clases.values())

# Graficar los datos
plt.bar(clases, conteos, color=['blue', 'green'])  # Colores opcionales
plt.xlabel('Clase')
plt.ylabel('Cantidad')
plt.title('Conteo de Clases')
plt.xticks(clases)  # Asegura que los ticks del eje x correspondan a las clases
plt.show()
```

#### Solo 492(0.172%) de las transacciones son fraudulentas. Eso significa que hay un gran desbalanceo con respecto a la variable Clase.

#### Existe un notorio desbalanceo entre las transacciones, por ello se debe realizar un balanceo de los datos, ya que con los datos actuales, la clasificación se sesgará y presentará una tendencia hacia la clase mayoritaria.

# Continuamos explorando la data

### Densidad de tiempo para las clases de transacciones de tarjetas de crédito.

```python
import matplotlib.pyplot as plt
import seaborn as sns

clase_0 = data.loc[data['Class'] == 0]['Time']
clase_1 = data.loc[data['Class'] == 1]['Time']

# Graficamos

sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
sns.kdeplot(clase_0, label='No fraude', fill=True)
sns.kdeplot(clase_1, label='Fraude', fill=True)
plt.title('Tiempo para las clases de transacciones')
plt.xlabel('Time [s]')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

#### Podemos observar que las transacciones fraudulentas tienen una distribución más uniforme que las transacciones válidas, se distribuyen más equitativamente en el tiempo, incluidos los bajos tiempos de transacción real en la zona horaria de Europa.

# Correlación de características

```python
plt.figure(figsize=(14,14))
plt.title('Correlación de caracteristicas')
corr = data.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns,linewidths=.1,cmap='Reds')
plt.show()
```

#### No existe una correlación entre las características(v1-v28). Existen ciertas correlaciones entre algunas de estas características y el tiempo(correlación inversa con v3), y la cantidad(correlación directa con v7 y v20), y correlación inversa con v1 y v5.

```python
# Vamos a visualizar diagramas de caja de cada variable en relación con la variable objetivo('class').

fig, axes = plt.subplots(nrows=6, ncols=5, figsize=(30,40))
fig.suptitle('Diagrama de caja de variables vs Clase', size=18, y=0.9, fontweight='bold')


fila = 0
columna = 0

for i in data.columns[:-1]:
    sns.boxplot(ax=axes[fila,columna], data=data, x='Class', hue='Class', y=i, palette='viridis', legend=False)
    axes[fila, columna].set_title('f{i} distribución')
    columna = columna + 1
    if columna == 5:
        columna = 0
        fila += 1

```

```python
# Clasificación de clases '0 = Normal y 1 = Fraude'

normal = data[data['Class'] == 0]
fraude = data[data['Class'] == 1]
```

```python
f, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
f.suptitle('Monto de transacción por clase')
bins = 50
ax1.hist(fraude.Amount, bins = bins)
ax1.set_title('Fraude')
ax2.hist(normal.Amount, bins = bins)
ax2.set_title('Normal')
plt.xlabel('Monto')
plt.ylabel('Número de transacciones')
plt.xlim((0, 20000))
plt.yscale('log')
plt.show()
```

```python
f, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
f.suptitle('Time of transaction vs Amount by class')
ax1.scatter(fraude.Time, fraude.Amount)
ax1.set_title('Fraude')
ax2.scatter(normal.Time, normal.Amount)
ax2.set_title('Normal')
plt.xlabel('Tiempo en segundos')
plt.ylabel('Monto')
plt.show()
```
