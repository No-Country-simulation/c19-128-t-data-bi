```python
# Convertiremos las clases a valores númericos para construir correlation matrix

from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
data['class_numerica'] = le.fit_transform(data['calss'])
```

```python
print(data['class_numerica'].unique())
```

```python
import matplotlib.pyplot as plt
# Seleccionar las características numéricas para el análisis de correlación
columnas_numericas = ['total_fpackets', 'total_bpackets', 'total_fpktl', 'total_bpktl', 'mean_fpktl', 'mean_bpktl', 'class_numerica']
df_numeric = data[columnas_numericas]

# Calcular la matriz de correlación
matriz_correlacion = df_numeric.corr()

# Visualizar la matriz de correlación como un mapa de calor
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de correlación')
plt.show()
```

```python
# Histograma para la duración

plt.hist(data['duration'], bins=30, edgecolor='black')
plt.xlabel('Duración')
plt.ylabel('Frecuencia')
plt.title('Histograma de duración')
plt.show()
```
